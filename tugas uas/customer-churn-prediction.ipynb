#### <b>Apa itu Churn Pelanggan?</b>
<span style="font-size:16px;">Churn pelanggan didefinisikan sebagai ketika pelanggan atau pelanggan berhenti berbisnis dengan suatu perusahaan atau layanan.</span>

<span style="font-size:16px;">Pelanggan di industri telekomunikasi dapat memilih dari berbagai penyedia layanan dan secara aktif beralih dari satu penyedia ke penyedia lainnya. Bisnis telekomunikasi memiliki tingkat churn tahunan sebesar 15-25 persen di pasar yang sangat kompetitif ini.</span>

<span style="font-size:16px;">Pertahanan pelanggan yang individualisasi sulit dilakukan karena kebanyakan perusahaan memiliki banyak pelanggan dan tidak dapat menghabiskan banyak waktu untuk setiap pelanggan. Biaya akan terlalu besar, melebihi pendapatan tambahan yang diperoleh. Namun, jika suatu perusahaan dapat memprediksi pelanggan mana yang kemungkinan besar akan meninggalkan di masa depan, perusahaan dapat fokus pada upaya retensi pelanggan hanya pada pelanggan "berisiko tinggi" ini. Tujuan utamanya adalah untuk memperluas jangkauan dan mendapatkan lebih banyak loyalitas pelanggan. Inti kesuksesan di pasar ini terletak pada pelanggan itu sendiri.</span>

<span style="font-size:16px;">Churn pelanggan adalah metrik yang penting karena lebih murah untuk mempertahankan pelanggan yang ada daripada mengakuisisi pelanggan baru.</span>

<a id="churn"></a>
<a id = "3" ></a>

<span style="font-size:16px;"><b>Untuk mengurangi churn pelanggan, perusahaan telekomunikasi perlu memprediksi pelanggan mana yang berisiko tinggi untuk churn.</b></span>

<span style="font-size:16px;">Untuk mendeteksi tanda-tanda awal churn potensial, pertama-tama harus dikembangkan pandangan menyeluruh tentang pelanggan dan interaksi mereka di berbagai saluran, termasuk kunjungan toko/cabang, riwayat pembelian produk, panggilan layanan pelanggan, transaksi berbasis web, dan interaksi media sosial, hanya beberapa di antaranya.</span>

<span style="font-size:16px;">Dengan mengatasi churn, perusahaan-perusahaan ini tidak hanya dapat mempertahankan posisi pasar mereka, tetapi juga tumbuh dan berkembang. Semakin banyak pelanggan yang mereka miliki dalam jaringan mereka, semakin rendah biaya inisiasi dan semakin besar keuntungannya. Oleh karena itu, fokus utama perusahaan untuk mencapai kesuksesan adalah mengurangi perpindahan pelanggan dan mengimplementasikan strategi retensi yang efektif.</span>

```python
def object_to_int(dataframe_series):
    if dataframe_series.dtype=='object':
        dataframe_series = LabelEncoder().fit_transform(dataframe_series)
    return dataframe_series
```

```python
df = df.apply(lambda x: object_to_int(x))
df.head(20)
```

```python
plt.figure(figsize=(14,7))
df.corr()['Churn'].sort_values(ascending = False)
```

```python
# Menghitung korelasi antara fitur-fitur dengan kolom 'Churn' dan mengurutkannya secara menurun
correlation = df.corr()['Churn'].sort_values(ascending=False)

# Mengambil 4 fitur dengan korelasi tertinggi kecuali 'Churn'
top_4_features = correlation[1:5].index.tolist()

# Menampilkan fitur-fitur yang akan digunakan untuk training model
print("Fitur-fitur untuk training model:")
print(top_4_features)
```

```python
# Mengambil fitur-fitur dari dataset ke dalam variabel X
X = df[top_4_features]

# Mengambil target (label) dari dataset ke dalam variabel y
y = df['Churn'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 42)
```

#### <b> KNN</b>
<a id = "knn" ></a>

```python
import pickle
from sklearn.neighbors import KNeighborsClassifier

# Membuat objek model KNN
knn_model = KNeighborsClassifier(n_neighbors=11)

# Melatih model dengan data latih yang telah di-oversampling
knn_model.fit(X_train, y_train)

# Memprediksi target menggunakan data uji
predicted_y = knn_model.predict(X_test)

# Menghitung akurasi model
accuracy_knn = accuracy_score(y_test, predicted_y)

print("KNN accuracy:", accuracy_knn)

# Simpan model ke dalam file pkl
with open('knn_model.pkl', 'wb') as file:
    pickle.dump(knn_model, file)
```

```python
import pickle

# Load model dari file pkl
with open('knn_model.pkl', 'rb') as file:
    knn_model = pickle.load(file)

# Membuat input data baru untuk pengujian
new_data = [[58.50, 0, 0, 3],    # Contoh data 1
            [61.55, 0, 0, 2],    # Contoh data 2
            [69.95, 1, 0, 2],    # Contoh data 3
            [73.60, 0, 0, 2],    # Contoh data 4
            [70.25, 1, 1, 0]]    # Contoh data 5 (dengan prediksi 1)

# Melakukan prediksi menggunakan model KNN
predictions = knn_model.predict(new_data)

# Menampilkan hasil prediksi
for data, prediction in zip(new_data, predictions):
    print("Input Data:", data)
    print("Prediction:", prediction)
    print()
```

```python
print(classification_report(y_test, predicted_y))
```

```python
# Membuat objek model Random Forest Classifier dengan parameter yang telah ditentukan
model_rf = RandomForestClassifier(n_estimators=500 , oob_score = True, n_jobs = -1,
                                  random_state =50, max_features = "auto",
                                  max_leaf_nodes = 30)

# Melatih model menggunakan data latih
model_rf.fit(X_train, y_train)

# Melakukan prediksi menggunakan data uji
prediction_test = model_rf.predict(X_test)

# Menghitung akurasi dengan membandingkan prediksi dengan nilai sebenarnya
accuracy = metrics.accuracy_score(y_test, prediction_test)
print(accuracy)
```
