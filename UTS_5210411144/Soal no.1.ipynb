Kode dibawah digunakan untuk mengambil library sklearn untuk memgatur label encoder, KNN, memisah data, dan tes skor akurasi.

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

Kode ini digunakan untuk mengubah label string ke label angka/numerik.

```python
# Label encode the response variable
le = LabelEncoder()
print(dataku)

for col in dataku.columns.values:
  # Encoding pada variabel kategorial
  if dataku[col].dtypes == 'object':
    data=dataku[col].append(dataku[col])
    le.fit(data.values)
    dataku[col]=le.transform(dataku[col])

dataku.head(10)
```

Seperti kode di atas, kode ini digunakan untuk mengisi variable x dan y dengan feature matrix dan response vector.

```python
x = dataku[dataku.columns[:-1]]
y = dataku[dataku.columns[-1]]
```

Memisah data menjadi data training dan data testing

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

Membuat KNN model dengan K=3 dan K=4 untuk melihat perbedaan.

```python
knn_3 = KNeighborsClassifier(n_neighbors=3)
knn_4 = KNeighborsClassifier(n_neighbors=4)
```

Melati model KNN dengan data training.

```python
knn_3.fit(x_train, y_train)
knn_4.fit(x_train, y_train)
```

Membuat prediksi pada data testing.

```python
y_pred_3 = knn_3.predict(x_test)
y_pred_4 = knn_4.predict(x_test)
```

Mengevaluasi akurasi model KNN yg sudah diuji.

```python
accuracy_3 = accuracy_score(y_test, y_pred_3)
accuracy_4 = accuracy_score(y_test, y_pred_4)
print("Accuracy of KNN with K=3: {:.2f}%".format(accuracy_3 * 100))
print("Accuracy of KNN with K=4: {:.2f}%".format(accuracy_4 * 100))
```

Dapat terlihat bahwa Akurasi KNN dengan K=3 dan K=4 yang memiliki hasil akurasi yang sama atau seimbang
