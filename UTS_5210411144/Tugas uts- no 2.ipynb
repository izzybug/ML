**DBSCAN (Density-Based Spatial Clustering of Applications with Noise)** adalah salah satu algoritma clustering yang mengelompokkan data berdasarkan kerapatan atau kepadatan data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import adjusted_rand_score, silhouette_score
```

```python
# Load dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
breast_cancer_df = pd.read_csv(url, header=None)
```

```python
# Select two features for clustering
X = breast_cancer_df.iloc[:, 2:4].values
```

```python
# Standardize the features
X = StandardScaler().fit_transform(X)
```

```python
# Perform DBSCAN clustering
dbscan = DBSCAN(eps=0.4, min_samples=5)
y_pred = dbscan.fit_predict(X)
```

```python
# Plot the clustering results
plt.scatter(X[:,0], X[:,1], c=y_pred, cmap='Paired')
plt.title('DBSCAN Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

**Adjusted Rand Index (ARI)**
ARI adalah ukuran yang digunakan untuk mengukur seberapa baik pengelompokan model dibandingkan dengan pengelompokan yang benar atau "ground truth". Nilai ARI berkisar antara -1 hingga 1, di mana 1 menunjukkan kesesuaian sempurna antara kedua pengelompokan dan 0 menunjukkan ketidaksamaan antara kedua pengelompokan secara acak.

**Silhouette Coefficient**
Silhouette Coefficient adalah ukuran yang mengukur seberapa baik suatu objek cocok dengan pengelompokannya sendiri dibandingkan dengan pengelompokan lain. Nilai Silhouette Coefficient berkisar antara -1 hingga 1, di mana nilai 1 menunjukkan objek yang sangat cocok dengan pengelompokannya dan nilai -1 menunjukkan objek yang sangat tidak cocok dengan pengelompokannya.

```python
# Calculate ARI
ari = adjusted_rand_score(breast_cancer_df.iloc[:, 1].values, y_pred)
```

```python
# Calculate Silhouette Coefficient
silhouette = silhouette_score(X, y_pred)
```

```python
print('Silhouette Coefficient:', silhouette)
print('Adjusted Rand Index:', ari)
```
